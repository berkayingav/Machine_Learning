GRADIENT DESCENT

```python
# maliyet cost hesaplama fonksiyonu

def cal_cost(W,X,y):
    """
    W = Katsayı vektörü -> (p,1)
    X = Girdi vektörü -> (n,p)
    y = Gerçek y'ler ->  (n,1)
    
    n = toplam veri adeti (satır sayısı)
    p = X içindeki değişken adedi sayısı (sütun sayısı)
    
    Dönüş:
    Cost = hesaplanan maliyet değeri
    """
    
    n = len(y)
    
    #önce tahmini hesapla  -> y_pred = XW
    
    tahmin = X.dot(W)
    
    # cost function değerini hesapla 
    cost = (1/2*n)* np.sum(np.square(tahmin-y))
    
    #cost'u return et
    
    return cost
```

```python
# Gradient Descent fonksiyonu
def gradient_descent(X,y,W,learning_rate=0.01,iterations=100):
    """
    Gradient descent uygulama fonksiyonu 
    Parametreler:
    X = X matrisi bias unit eklenmiş hali
    y = y vektörü
    W = Katsayı vektörü
    learning_rate = a = öğrenme katsayısı
    iterations  = toplam döngü sayısı
    
    Dönüş:
    W vektörünün son hali
    Maliyet listesi (cost history)
    W vektörünün listesi (weight history)
    """
    
    n = len(y)
    cost_history = np.zeros(iterations)
    w_history= np.zeros((iterations,2))
    
    for it in range(iterations):
        
        tahmin = np.dot(X,W)
        
        W = W - (1/n) * learning_rate * (X.T.dot((tahmin-y)))
        
        w_history[it,:] = W.T
        cost_history[it] = cal_cost(W,X,y)
        
    return W,cost_history, w_history
```

```python
lr = 0.01  # learning rate

n_iter = 1000  # İterasyon sayısı

#başlangıç için W vektörü

W = np.random.randn(2,1)

#X vektörüne bias columns ekle

X_b = np.c_[np.ones((len(X),1)),X]

# Gradient Descent çalıştır

W,cost_history,w_history = gradient_descent(X_b,y,W,lr,n_iter)

print('W0: {:0.3f}'.format(W[0][0]))
print('W1: {:0.3f}'.format(W[1][0]))
print('Final cost (MSE): {:0.3f}'.format(cost_history[-1]))
```

```python
# İterasyonları çizelim

fig,ax = plt.subplots(figsize=(12,8))

ax.set_ylabel('J(W)')
ax.set_xlabel('Iterations')

_=ax.plot(range(n_iter),cost_history,'b.')
plt.show()
```

```python
#Gradient Descenti Çizen Fonksiyon

def plot_GD(n_iter,lr,ax,ax1=None):
    """
    n_iter = iterasyon sayısı
    lr = learning rate
    ax = gradient descent çizmek için axis
    ax1 = cost history vs gradient descent için axis
    """
    _=ax.plot(X,y,'b.')
    W = np.random.randn(2,1)
    
    tr = 0.1
    cost_history = np.zeros(n_iter)
    
    for i in range(n_iter):
        pred_prev = X_b.dot(W)
        
        W,h,_= gradient_descent(X_b,y,W,lr,1)
        
        pred = X_b.dot(W)
        
        cost_history[i] = h[0]
        
        if((i % 25 == 0)):
            _= ax.plot(X,pred,'r-',alpha=tr)
            if tr < 0.8:
                tr = tr + 0.2
                
    if not ax1 == None:
        _= ax1.plot(range(n_iter),cost_history,'b.')
            
    
```

```python
fig = plt.figure(figsize=(30,25))

fig.subplots_adjust(hspace=0.4, wspace=0.4)

#iterasyon ve learning rate listesi

it_lr = [(2000,0.001),(500,0.01),(200,0.05),(100,0.1)]

count = 0

for n_iter, lr in it_lr:
    count += 1
    
    ax =fig.add_subplot(4,2,count)
    count +=1
    
    ax1 = fig.add_subplot(4,2,count)

    
    ax.set_title("lr:{}".format(lr))
    ax1.set_title("iterations:{}".format(n_iter))
    
    plot_GD(n_iter, lr, ax, ax1)
```

STOCHASTIC GRADIENT DESCENT

```python
def stochastic_gradient_descent(X,y,W,learning_rate=0.01,iterations=100):
    """
    X = Bias ünit eklenmiş hali X vektörünün
    y = y vekötür
    W = katsayı vektörü
    learning_rate = alpha = öğrenme katsayısı
    iterations = toplam döngü sayısı
    Dönüş:
    *W vektörünün son hali
    *Maliyet listesi (cost history)
    *W vektörünün listesi (weight history) 
    """
    
    n = len(y)
    cost_history=np.zeros(iterations)
    
    for it in range(iterations):
        cost = 0
        
        for i in range(n):
            
            rand_ind = np.random.randint(0,n) # randint belirtilen aralıkta dahil olmayan anlamına gelir yani n dahil değil
            
            X_i = X[rand_ind,:].reshape(1,X.shape[1])  #.reshape amacı satır matrisine çevirmektir. Yani aslında bir satır alınır burada
            y_i = y[rand_ind].reshape(1,-1)
            
            # Tek X_i için tahmin
            
            tahmin = X_i.dot(W)
            
            # Tek X_i için katsayı değişimi (formül)
            
            W = W - (1/n) * learning_rate * (X_i.T.dot((tahmin - y_i)))
            
            # hesaplanan costu cost değişkenine ekle 
            
            cost += cal_cost(W,X_i,y_i)
        # Bu iterasyon için cost değerini cost history'e ekle
        cost_history[it]=cost
        
    return W , cost_history
```

```python
lr = 0.5

n_iter = 50

W = np.random.randn(2,1)

X_b = np.c_[np.ones((len(X),1)),X]

W , cost_history = stochastic_gradient_descent(X_b,y,W,lr,n_iter)

print('W0: {:0.3f}'.format(W[0][0]))
print('W1: {:0.3f}'.format(W[1][0]))
print('Final cost/MSE: {:0.3f}'.format(cost_history[-1]))
```

Mini Bath Gradient Descent

```python
def minibatch_stochastic_gradient_descent(X,y,W,learning_rate=0.01,iterations=100,batch_size=20):
    """
    Parametreler:
    X = X matrisin (bias unit eklenmiş hali)
    y = y vektörü 
    W = w'lerden oluşmuş katsayı vektörü 
    learning_rate = alpha = öğrenme katsayısı
    iterations = toplam döngü sayısı
    Dönüş:
    * W vektörünün son hali
    * Maliyet listesi (cost history)
    * W vektörünün listesi (weight history)
    """
    
    n = len(y)
    cost_history = np.zeros(iterations)
    # toplamda kaç adet batch olacak
    n_batches = int(n/batch_size)
    
    for it in range(iterations):
        cost = 0
        # X ve y leri karıştıracağız rastgelelik için
        indices = np.random.permutation(n)  # 0 ile n-1 arasındaki sayıları karıştırarak dizi oluşturur n kadar
        X = X[indices]
        y = y[indices]
        
        # Her seferinde rastgale bir X_i alıp maliyeti hesaplayacağız
        
        for i in range(0,n,batch_size):
            
            X_i = X[i:i+batch_size]
            y_i = y[i:i+batch_size]
            
            # bias columns ekle
            
            #x_i =  np.c_[np.ones(len(X_i)),X_i]
            
            tahmin = X_i.dot(W)
            
            W = W - (1/n) * learning_rate * (X_i.T.dot((tahmin - y_i)))
            
            cost += cal_cost(W,X_i,y_i)
            
        cost_history[it] = cost
        
    return W , cost_history
        
```

```python
lr = 0.1

n_iter = 200 

W = np.random.randn(2,1)

X_b = np.c_[np.ones((len(X),1)),X]

W , cost_history = minibatch_stochastic_gradient_descent(X_b,y,W,lr,n_iter)

print('W0: {:0.3f}'.format(W[0][0]))
print('W1: {:0.3f}'.format(W[1][0]))
print('Final cost/MSE: {:0.3f}'.format(cost_history[-1]))
```

```python
# Grafik üzerinde görelim
fig,ax = plt.subplots(figsize=(10,8))

ax.set_ylabel('{J(W)}',rotation=0)
ax.set_xlabel('{iterations}')

_=ax.plot(range(n_iter),cost_history,'b.')
plt.show()
```
