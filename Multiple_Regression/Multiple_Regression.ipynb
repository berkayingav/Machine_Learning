OLS ANALÄ°ZÄ°

```python
#Ols iÃ§in datamÄ±zÄ± kopyalayalÄ±m
X_train_rfemodel = X_train_rfe.copy()
```

```python
# add constant iÃ§in beta0 iÃ§in gerektiÄŸinden 1'lerden oluÅŸan bir sÃ¼tun oluÅŸturucaz 

X_train_rfemodel = sm.add_constant(X_train_rfemodel)
```

```python
X_train_rfemodel
```

```python
X_train_rfemodel.columns
```

```python
#OLS Ã§alÄ±ÅŸtÄ±rÄ±yoruz
lr = sm.OLS(y_train,X_train_rfemodel).fit()
```

```python
print(lr.summary())
```

```python
# Burada rotor deÄŸeri 0.05 den bÃ¼yÃ¼k olduÄŸu iÃ§in Ã¶nemsiz statÃ¼sÃ¼ndedir.
#DolayÄ±sÄ±yla Ã¶nemsizdir ve onu Ã§Ä±karabiliriz.

X_train_rfemodel = X_train_rfemodel.drop(['rotor'],axis=1)
```

```python
# Ols yi tekrar Ã§alÄ±ÅŸtÄ±rÄ±caz sonra
```

```python
def train_ols(X,y):
    X = sm.add_constant(X)
    lr = sm.OLS(y,X).fit()
    print(lr.summary())
```

```python
#Åžimdi Ã§aÄŸÄ±ralÄ±m
train_ols(X_train_rfemodel,y_train)
```

```python
# GÃ¶rÃ¼ndÃ¼ÄŸÃ¼ Ã¼zere 0.05 ten bÃ¼yÃ¼k olduÄŸu iÃ§in dohcv'de Ã¶nemsizdir. Onuda Ã§Ä±karÄ±yoruz
X_train_rfemodel = X_train_rfemodel.drop(['dohcv'],axis=1)
```

```python
train_ols(X_train_rfemodel,y_train)
```

```python
# GÃ¶rÃ¼ldÃ¼ÄŸÃ¼ Ã¼zere five deÄŸiÅŸkenide Ã¶nemsiz. Onuda Ã§Ä±karacaÄŸÄ±z

X_train_rfemodel= X_train_rfemodel.drop(['five'],axis=1)
train_ols(X_train_rfemodel,y_train)
```
