OLS ANALİZİ

```python
#Ols için datamızı kopyalayalım
X_train_rfemodel = X_train_rfe.copy()
```

```python
# add constant için beta0 için gerektiğinden 1'lerden oluşan bir sütun oluşturucaz 

X_train_rfemodel = sm.add_constant(X_train_rfemodel)
```

```python
X_train_rfemodel
```

```python
X_train_rfemodel.columns
```

```python
#OLS çalıştırıyoruz
lr = sm.OLS(y_train,X_train_rfemodel).fit()
```

```python
print(lr.summary())
```

```python
# Burada rotor değeri 0.05 den büyük olduğu için önemsiz statüsündedir.
#Dolayısıyla önemsizdir ve onu çıkarabiliriz.

X_train_rfemodel = X_train_rfemodel.drop(['rotor'],axis=1)
```

```python
# Ols yi tekrar çalıştırıcaz sonra
```

```python
def train_ols(X,y):
    X = sm.add_constant(X)
    lr = sm.OLS(y,X).fit()
    print(lr.summary())
```

```python
#Şimdi çağıralım
train_ols(X_train_rfemodel,y_train)
```

```python
# Göründüğü üzere 0.05 ten büyük olduğu için dohcv'de önemsizdir. Onuda çıkarıyoruz
X_train_rfemodel = X_train_rfemodel.drop(['dohcv'],axis=1)
```

```python
train_ols(X_train_rfemodel,y_train)
```

```python
# Görüldüğü üzere five değişkenide önemsiz. Onuda çıkaracağız

X_train_rfemodel= X_train_rfemodel.drop(['five'],axis=1)
train_ols(X_train_rfemodel,y_train)
```
