TRAİN - TEST - SPLİT

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y ,test_size = 0.3 , random_state = 0 )
```

FEATURE SCALİNG

```python
cols = X_train.columns
cols
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)

X_test = sc.transform(X_test)


# Scaling işlemi boyutlar için kullanılır. Mesela proline 1000lerdeyken malic acid 1lerde. Bu yüzden proline daha önemli 
# algılanır. Dolayısıyla bunu önlemek için scaling işlemi yapılır.
```

```python
print(type(X_train))
print(type(X_test))   

# scalere verdiğimiz için numpy. ndarraya döndü. O yüzden tekrardan dataframeye çeviriyoruz.
```

```python
X_train = pd.DataFrame(X_train,columns=[cols])
X_test = pd.DataFrame(X_test,columns=[cols])
```

```python
X_train.head()
```

```python
print(type(X_train))
print(type(X_test))   
```

```python
X_test.head()
```

MODEL GELİŞTİRME

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
```

```python
# Modeli eğitelim

gnb.fit(X_train,y_train)
```

SONUÇ TAHMİNİ

```python
y_pred = gnb.predict(X_test)
y_pred
```

TAHMİN SONUCU ANALİZİ

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
# Sonuç %94 çıktı
```

Accuracy değerimiz çok iyi çıktı

Acaba overfitting veya underfittingmi var ? 

Overfitting = Eğer train data skoru çok yüksek  ama test data skoru çok düşük çıksaydı

Underfitting = Eğer train data skoru çok düşük ama test data skoru çok yüksek çıksaydı

test data çok yüksek çıktığına göre acaba underfittingmi var daha doğru bir sorudur.

```python
y_pred_train = gnb.predict(X_train)
y_pred_train
```

```python
accuracy_score(y_train,y_pred_train)
```
