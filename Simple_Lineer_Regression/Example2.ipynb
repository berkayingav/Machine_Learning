```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/Advertising.csv',index_col=0)
```

```python
df.describe()
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
data = df[['radio','sales']]

X = data['radio']

y = data['sales']
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data = data , x = X , y = y, c = 'Blue')

plt.title('radio-SATIS')
plt.xlabel('radio')
plt.ylabel('Satıs')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X.shape
```

```python
y.shape
```

```python
X = X.values.reshape(-1,1)
y = y.values.reshape(-1,1)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3 , random_state = 100)
```

```python
print("X_train size",X_train.shape)
```

```python
print("y_test size", y_train.shape)
```

```python
lr.fit(X_train,y_train)
```

```python
print("Value of İntercept :", lr.intercept_)
```

```python
print("Value of Slope coefficient :", lr.coef_)
```

```python
# y = B0 + B1X from the formula 
# y = 9.25 + 0.20X
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_pred.shape
```

```python
# Draw the grand truth ( the real data) and Prediction data
```

```python
#Real Data

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X_test, y_test , label = "Grand Truth" , c ='red')

# Prediction

ax.scatter(X_test, y_pred , label = "Prediction", c = 'blue')

plt.title('radio-SALES PREDİCTİON')
plt.xlabel('radio')
plt.ylabel('SALES')
plt.legend(loc = 'upper left')
plt.show()
```

```python
y_test[0:10]
```

```python
y_pred[0:10]
```

```python
# see the change at each prediction point

indexs = range(1,61)

# Real data -> Grand truth

fig,ax = plt.subplots(figsize=(12,8))
ax.plot(indexs,y_test,label ='Grand Truth',color='Red',linewidth=2)

# Prediction
ax.plot(indexs,y_pred,label = 'Prediction', color = 'Black' ,linewidth=2)

plt.title('Truth - Prediction')
plt.xlabel('Data Index')
plt.ylabel('SALES')
plt.legend(loc = 'upper left')
plt.show()
```

```python
# Draw the errors (Residual)

# Error = Residual -> y-y^ 

# Error = y_test - y_pred
```

```python
import numpy as np
```

```python
indexs = range(1,61)

fig,ax = plt.subplots(figsize=(12,8))
ax.plot(indexs , y_test-y_pred, label ='Residual', color ='Red')

# Zero plot

ax.plot(indexs , np.zeros(60),color='Blue')

plt.title('Residual Graph')
plt.xlabel('Data Index')
plt.ylabel('Sales Residual')
plt.show()
```

```python
#CHECKING THE MODEL ACCURACY
```

```python
# RMSE = Root Mean Squared Error 

# RMSE = The square root of the MSE 

# R^2 = coefficients of determination
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r_2 = r2_score(y_test,y_pred)
```

```python
print("R^2:",r_2)
```

```python
print("%R^2:",r_2*100)
```

```python
# calculate the RMSE

mse = mean_squared_error(y_test,y_pred)
```

```python
print("MSE:",mse)
```

```python
# calculate the RMSE
import math
```

```python
rmse = math.sqrt(mse)
```

```python
print("RMSE:",rmse)
```
