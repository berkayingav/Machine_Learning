```python
data = df[['TV','sales']]

#input X

X = data['TV']

#output y

y = data['sales']
```

```python
#grafik çiz

plt.figure(figsize=(12,8))
sns.scatterplot(data=data, x='TV', y='sales', color='red') 
#bir data istedi data'yı girdik. Ardından ise x ve y kısımlarımızı belirledik. x = X ve y = y  diye de yazabilirdik.
plt.title('TV-SATIŞ')
plt.xlabel('TV')
plt.ylabel('SATIŞ')
plt.show()

# Böylece datayı görselleştirmiş olduk.
```

```python
# ŞİMDİ MODELİ OLUŞTURACAĞIZ
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
# HAZIRLIK - BOYUT KONTROLÜ YAPICAZ
```

```python
#X ve y'in boyutlarına bakacağız
X.shape
y.shape
```

```python
# Şimdi ise yeniden boyutlandırmamız lazım
# boyutlandırırken 200 1 şeklinde boyutlandıracağız çünkü 200 satır bir sütunumuz var
X = X.values.reshape(-1,1) 
# bu yazım ne kadar boyutu varsa kabul et (-1) ve +1 sütun ekle demek
X.shape
y = y.values.reshape(-1,1)
y.shape
```

```python
#Toplam 200 verimizi test ve train olmak üzere ayıracağız
# %70 train data 
#  Train Data:  X_train , y_train

# %30 test data
# Test Data: X_test , y_test
```

```python
# scikit-Learn'dan import

from sklearn.model_selection import train_test_split
```

```python
# Train Data
# X_train
# y_train

# Test Data
# X_test
# y_test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 100) #%30 olduğu için 0.30
```

```python
print("X_train'in boyutu:", X_train.shape)
print("y_train'in boyutu:", y_train.shape)
print("X_test'in boyutu:", X_test.shape)
print("y_test'in boyutu:", y_test.shape)
```

```python
type(X_train)
```

```python
# LİNEER REGRESYON MODELİNİ ÇALIŞTIRMA
```

```python
# fit metodu ile train data üzerinden öğrenme yapıcaz
lr.fit(X_train, y_train)
```

```python
# KATSAYILARI HESAPLAYACAĞIZ
```

```python
# intercept -> beta 0

print("intercept değeri:",lr.intercept_)
```

```python
# slope (coefficient) eğim -> beta 1

print("slope (coefficient) eğim",lr.coef_)
```

```python
# y = B0 + B1X formülünden 
# y = 6.98966586 + 0.04649736X
```

```python
# TAHMİN YAP
```

```python
#Regresyon sayılarımızı hesapladığımıza göre şimdi predict tahmin yapıcaz y_pred
```

```python
#tahmin yap

y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_pred.shape
```

```python
# GERÇEK DATA VE TAHMİN DATASINI ÇİZDİRECEĞİZ
```

```python
# Gerçek label -> y_test
# Tahmin label -> y_pred
# input -> X_test
```

```python
# Gerçek data = Grand Truth
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X_test, y_test, label='Grand Truth', color='Red')

# Tahmin -> Prediction

ax.scatter(X_test, y_pred, label='Prediction', color='Black')


plt.title('TV-SALES PREDİCTİON')
plt.xlabel('TV')
plt.ylabel('SALES')
plt.legend(loc = 'upper left')
plt.show()
```

```python
# İlk 10 gerçek y değerlerini görelim

y_test[0:10]
```

```python
# ilk 10 tahmin y değerlerini görelim

y_pred[0:10]
```

```python
# Her bir tahmin noktasındaki değişimi görelim

indexler = range(1,61)

# Gerçek data -> Grand truth

fig,ax = plt.subplots(figsize=(12,8))
ax.plot(indexler,y_test,label ='Grand Truth',color='Red',linewidth=2)

# Tahmin -> Prediction
ax.plot(indexler,y_pred,label = 'Prediction', color = 'Black' ,linewidth=2)

plt.title('Gerçek - Prediction')
plt.xlabel('Data Index')
plt.ylabel('SALES')
plt.legend(loc = 'upper left')
plt.show()
```

```python
# HATALARI ÇİZDİRİCEZ 

# hata = Residual -> y-y^ 

# hata = y_test - y_pred
```

```python
import numpy as np
```

```python
indexler = range(1,61)

# Residuals -> Hatalar 

fig,ax = plt.subplots(figsize=(12,8))
ax.plot(indexler,y_test - y_pred,label ='Residuals',color='Red',linewidth=2)

# Sıfır doğrusunu çizin
ax.plot(indexler,np.zeros(60),color = 'Black' )

plt.title('Hatalar (Residuals)')
plt.xlabel('Data Index')
plt.ylabel('SALES')
plt.legend(loc = 'upper left')
plt.show()
```

```python
# MODEL DOĞRULUĞUNU KONTROL ETME (RMSE VE R^2)
```

```python
# RMSE = Root Mean Squared Error 

# RMSE = MSE'nin kareköküdür aslında

# R^2 = coefficients of determination
```

```python
#sklearn import

from sklearn.metrics import r2_score, mean_squared_error
```

```python
# R kareyi hesaplayalım

r_2 = r2_score(y_test,y_pred)
```

```python
#R^2 yi yazdır

print('R2:',r_2)
```

```python
#R^2 yi yazdır yüzde olarak

print('%R2:',r_2*100)
```

```python
# MSE --> RMSE

mse = mean_squared_error(y_test,y_pred)
```

```python
#mse yi yazdır

print('MSE:',mse)
```

```python
#RMSE

import math

rmse = math.sqrt(mse)
```

```python
#RMSE'yi yazdır

print('RMSE:',rmse)
```

# YORUMLAR 

R^2  = %59 YANİ YAKLAŞIK %60 ÇIKTI 

BU DEMEK BİZİM MODELİMİZ DATAYI %60 ORANINDA TEMSİL EDİYOR DEMEKTİR.

RMSE = 2.82
